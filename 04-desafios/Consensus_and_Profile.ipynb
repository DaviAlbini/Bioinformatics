```python
# seqs = """
# >Rosalind_1
# ATCCAGCT
# >Rosalind_2
# GGGCAACT
# >Rosalind_3
# ATGGATCT
# >Rosalind_4
# AAGCAACC
# >Rosalind_5
# TTGGAACT
# >Rosalind_6
# ATGCCATT
# >Rosalind_7
# ATGGCACT
# """

seqs = open('/content/rosalind_cons.txt', 'r').read()

genes = {}

entries = seqs.strip().split('>')
entries = [entry for entry in entries if entry]

for entry in entries:

  lines = entry.split('\n')

  gene_id = lines[:1]
  gene_id = ''.join(lines[:1]).strip()
  sequence = ''.join(lines[1:]).strip()
  genes[gene_id] = sequence

genes
```

```python
nucleic_acids = ['A', 'C', 'G', 'T']
profile = {}

seq_length = len(list(genes.values())[0])

for i in range(seq_length):
    for nucleotide in nucleic_acids:
        profile[(nucleotide, i)] = 0

for gene_id, sequence in genes.items():
    for i in range(seq_length):
        nucleotide = sequence[i]
        if nucleotide in nucleic_acids:
            profile[(nucleotide, i)] += 1

consensus_seq = ''
for i in range(seq_length):
    max_count = 0
    consensus_nucleotide = ''
    for nucleotide in nucleic_acids:
        count = profile[(nucleotide, i)]
        if count > max_count:
            max_count = count
            consensus_nucleotide = nucleotide
    consensus_seq += consensus_nucleotide

print(consensus_seq)

for nucleotide in nucleic_acids:
    print(f"{nucleotide}: ", end="")
    for i in range(seq_length):
        print(profile[(nucleotide, i)], end=" ")
    print()
```
