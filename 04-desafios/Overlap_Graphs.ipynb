```python
from io import StringIO

# Sample FASTA data from the user's message
# fasta_data = """>Rosalind_0498
# AAATAAA
# >Rosalind_2391
# AAATTTT
# >Rosalind_2323
# TTTTCCC
# >Rosalind_0442
# AAATCCC
# >Rosalind_5013
# GGGTGGG"""

fasta_data = open("/content/rosalind_grph.txt", "r").read()

# Function to parse FASTA format
def parse_fasta(fasta_str):
    fasta_io = StringIO(fasta_str)
    sequences = {}
    label = None
    for line in fasta_io:
        line = line.strip()
        if line.startswith(">"):
            label = line[1:]
            sequences[label] = ""
        else:
            sequences[label] += line
    return sequences

# Function to construct overlap graph O3
def construct_overlap_graph(sequences, k=3):
    adjacency_list = []
    for s_label, s_seq in sequences.items():
        s_suffix = s_seq[-k:]
        for t_label, t_seq in sequences.items():
            if s_label != t_label and t_seq.startswith(s_suffix):
                adjacency_list.append((s_label, t_label))
    return adjacency_list

# Parse the FASTA data
sequences = parse_fasta(fasta_data)

# Construct the overlap graph with k=3
overlap_graph = construct_overlap_graph(sequences, k=3)

# Print the adjacency list
for edge in overlap_graph:
    print(f"{edge[0]} {edge[1]}")
```
